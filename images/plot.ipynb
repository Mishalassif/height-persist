```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
data_guitar = np.genfromtxt('guitar_0155_red_pi.csv', skip_header=26)
print(data_guitar)
data_night = np.genfromtxt('night_stand_0200_red_pi.csv', skip_header=26)
print(data_night)
data_chair = np.genfromtxt('chair_0889_red_pi.csv', skip_header=26)
```

```python
data_night = np.reshape(data_night, (40,20,20))
k=5
plt.axis('off')
f, axs = plt.subplots(k,k,figsize=(10,10))
for i in range(k):
    for j in range(k):
        plt.subplot(k,k,i*k+j+1)
        plt.axis('off')
        plt.imshow(data_night[int((i*k+j)*40/(k*k))])
plt.subplots_adjust(wspace=0, hspace=0.1)
plt.savefig('night.png')
```

```python
data_guitar = np.reshape(data_guitar, (40,20,20))
k=5
plt.axis('off')
f, axs = plt.subplots(k,k,figsize=(10,10))
for i in range(k):
    for j in range(k):
        plt.subplot(k,k,i*k+j+1)
        plt.axis('off')
        plt.imshow(data_guitar[int((i*k+j)*40/(k*k))])
plt.subplots_adjust(wspace=0, hspace=0.1)
plt.savefig('guitar.png')
```
